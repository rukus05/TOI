```python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}

df = pd.DataFrame(data)

# Specify columns to check for changes
columns_to_check = ['Name', 'City']

# Iterate through rows using itertuples
for current_row, next_row in zip(df.itertuples(), df.itertuples()):
    current_values = [getattr(current_row, col) for col in columns_to_check]
    next_values = [getattr(next_row, col) for col in columns_to_check]

    # Compare values of specified columns for the current and next rows
    changes = [col for col, current, next in zip(columns_to_check, current_values, next_values) if current != next]

    if changes:
        print(f"Changes in row {current_row.Index} for columns {changes}:")
        print(df.loc[current_row.Index:current_row.Index + 1, columns_to_check])
        print("\n")
```
